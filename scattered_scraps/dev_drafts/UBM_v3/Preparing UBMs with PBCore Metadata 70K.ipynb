```python
import os
import random
import xmltodict
import pandas as pd
```

```python
### Creating list of names to exclude from UBM

label_data = pd.read_csv("/Volumes/McLaughlin-6TB-1/Dropbox/aapb-data/AAPB_corpus_metadata/AAPB_speech_labels_full_corpus_170414_2_sec_segments.csv")

formal_names = sorted(list(set(list(label_data['Value']))))
formal_names.remove('Classical Music')
formal_names.remove('NPR embedded feed')
formal_names.remove('audience clapping')
formal_names.remove('Spanish Language')
formal_names.remove('PRI embedded feed')
formal_names.remove('Dog Bark')
formal_names.remove('applause')

names = []

for name in formal_names:
    names.append(name)
    forward_order=' '.join([item.strip() for item in (name.split(', ')[::-1])])
    names.append(forward_order)
    names.append(forward_order.replace('.','').replace(',',''))


names = sorted(list(set(names)))
names
```

```python
## Getting list of PBCore filenames

os.chdir('/Users/mclaugh/Desktop/PBCore_Master_Set_70K')
filenames=os.listdir('./')
```

```python
## Checking each PBCore file for each name on our list and creating UBM blacklist

black_list=[]

for filename in filenames:
    with open(filename) as fi:
        body = fi.read().lower()
        for name in names:
            if name.lower() in body:
                print(name+" is in "+filename)
                black_list.append(filename)
                break
```

```python
with open('../UBM_black_list.txt','w') as fo:
    for item in black_list:
        fo.write(item+'\n')
```
